```python
# Create data            gràfica de (despeses de la federació) X (pobresa)en la zona 

x = conti.fed_spend      # poso la columna fed_spend  (despeses de la federació)a les X
y = conti.poverty        # poso poverty(pobresa) a la columna  de les Y
colors = 'Blue'           
area = np.pi*5             

plt.axis([-10, 100, 0, 60])  # em diu els valors que tindran els eixos de la gràfica de -10 a 100 el eix x, i de 0 a 60 l'eix y

# Plot
plt.scatter(x, y, s=area, c=colors, alpha=0.4, edgecolors='black')    # tipus de gràfica scatter

plt.title('Federal Spending vs Poverty by County') # títol de la gràfica
plt.ylabel('Federal Spending per Capita')          # nom de l'eix y
plt.xlabel('Poverty Rate (Percent)')               # nom de l'eix x
plt.show()                                         # mostra la gràfica
```

```python
fig = plt.figure(figsize=(15,4))  # em diu quan ocupara a la part de les x i quan a la part de les y de la gràfica 
                                  # quan més gran és el valor més ample o alta és la gràfica
ax1 = fig.add_subplot(1, 3, 1)

# Create data
x = conti.fed_spend         # columna fed_spend
y = conti.poverty           # columna poverty
colors = 'Blue'             # li indica a la var colors el color en que es veuran els punts
area = np.pi*5              # el 5 que li indiquem quan multipliquem pi ens donarà el tamany dels punts de la gráfica

plt.axis([0, 100, 0, 60])    # fin aquesta linia només fa la gràfica buida , valors dels eixos, x de 0 a 100 i y de 0 a 60

# Plot                       # tipus de gràfica scatter

ax1.scatter(x, y, s=area, c=colors, alpha=0.4, edgecolors='black')  # alpha ens dona La nitidesa, si augmento el 0.4 es veu més borrós

plt.title('Conti Dataset')                    # titol grafica
plt.ylabel('Federal Spending per Capita')     # nom eix y despeses federals per càpita
plt.xlabel('Poverty Rate (Percent)')          # nom eix x tassa de pobresa en porcentatge

```

```python
# Matrix Plots                                
                                              # demano fer un pairplot que dona un conjunt de gràfiques una matriu totes
                                              #de les columnes númeriques del dataset X per totes les columnes
        
sns.pairplot(conti, diag_kind='hist', plot_kws={'alpha': 0.2})      # {'alpha': 0.4} dona la nitidesa de les gràfiques
                                                                    
```

```python
fig = plt.figure(figsize=(10,8))     #tipus de grafica histogrames

ax1 = fig.add_subplot(2, 2, 1)       # al afegir el subplot reparteix l espai per disposar  les 4 grafiques

ax1.hist(conti['multiunit'], bins=25)  # histograma del al taula multiunit, poblacio a la zona al 2010 per multunitat
plt.title('Conti- 2010 Population')    #poblacio a la zona al 2010 
plt.ylabel('Frequency')
plt.xlabel('multi unit (%)')         #multiunitat en percentatge

ax2 = fig.add_subplot(2, 2, 2)

ax2.hist(conti['income'], bins=25)   # histograma de la columna income, poblacio a la zona al 2010 per ingresos per cápita

plt.title('2010 Conti Population')
plt.ylabel('Frequency')
plt.xlabel('Per Capita Income')   #ingressos per càpita de la columna income

ax3 = fig.add_subplot(2, 2, 3)

ax3.hist(conti['homeownership'], bins=25)  #propietat de vivenda, a la columna homeownership
plt.title('2010 Conti Population')         #població a la zona al 2010
plt.ylabel('Frequency')                    # poblacio a la zona al 2010 per propietaris de casa
plt.xlabel('Homeownership (%)')

ax4 = fig.add_subplot(2, 2, 4)

ax4.hist(conti['med_income'], bins=25)   #columna 'med_income'

plt.title('2010 Conti Population')  #poblacio a la zona al 2010 per  la mitja per casa
plt.ylabel('Frequency')
plt.xlabel('Median Household Imcome') # mitjana  

plt.tight_layout()
```
